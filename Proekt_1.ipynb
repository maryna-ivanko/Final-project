## 1. Удалите дубликаты и неактуальные столбцы.
   

Используем код, который заполняет пропущенные значения в столбцах City и Level of Deutsch значением, которое чаще всего встречается для данного Contact Name, чтобы не потерять данные о городах и уровне языка.


```python
mode_values_city = (
    deals_df
    .groupby('Contact Name')['City']
    .agg(lambda x: (
        x.mode()[0]
        if not x.mode().empty
        else None
    ))
)

deals_df['City'] = deals_df['Contact Name'].map(mode_values_city)
```

```python
mode_values_level = (
    deals_df
    .groupby('Contact Name')['Level of Deutsch']
    .agg(
        lambda x: (
            x.mode()[0]
            if not x.mode().empty
            else None
      ))
)

deals_df['Level of Deutsch'] = (
    deals_df['Contact Name']
    .map(mode_values_level)
)
```

### Обработка столбца с валютой

Приведение столбцов к единому виду, очистка от символов валют

```python
print(deals_df["Offer Total Amount"].unique().tolist())
print(deals_df["Initial Amount Paid"].unique().tolist())
```

```python
deals_df['Initial Amount Paid'] = (
    deals_df['Initial Amount Paid'].replace({9: 0})
)
```

*Очищаем и преобразуем данные в колонках с денежными значениями (Initial Amount Paid, Offer Total Amount), чтобы привести их к числовому формату (float) и стандартизировать валюту (евро).*

```python
def clean_currency(value):
    if isinstance(value, str):
        value = value.replace("$", "").replace("€", "").strip()
        value = value.replace("\xa0", "").replace(",", ".")
        value = value.replace(".", "", value.count(".") - 1)
        return float(value)
    elif isinstance(value, (int, float)):
        return float(value)
    return 0

deals_df["Initial Amount Paid"] = (
    deals_df["Initial Amount Paid"]
    .apply(clean_currency)
    .fillna(0)
)
deals_df["Offer Total Amount"] = (
    deals_df["Offer Total Amount"]
    .apply(clean_currency)
    .fillna(0)
)
currency_columns = ["Initial Amount Paid", "Offer Total Amount"]

for col in currency_columns:
    deals_df[col] = deals_df[col].apply(clean_currency)

deals_df.rename(
    columns={col: f"{col} (€)" for col in currency_columns},
    inplace=True
)
```

*Подсчитаем количество строк, где Offer Total Amount (€) меньше, чем Initial Amount Paid (€).Такие значения есть, можно сделать вывод, что данные были поменяны местами. Создаем маску для выбора этих строк. Меняем значения в этих строках местами.*

*При анализе столбцов "Offer Total Amount (€)", "Product", "Education Type" было виявлено, что если "Product" равен "UX/UI Design" и значение в столбце "Offer Total Amount (€)" равно 11000, то в соответствующих строках столбца "Education Type" будет значение "Morning".*

*Для строк, где "Offer Total Amount (€)" равен 11500, значение в столбце "Education Type" может быть только "Morning".*

```python
count = (
    deals_df["Offer Total Amount (€)"]
    < deals_df["Initial Amount Paid (€)"]
).sum()

print(
    f"Количество строк, где Offer Total Amount < Initial Amount Paid: {count}"
)
```

```python
mask_amount = (
    deals_df["Offer Total Amount (€)"]
    < deals_df["Initial Amount Paid (€)"]
)
deals_df.loc[mask_amount, ["Offer Total Amount (€)", "Initial Amount Paid (€)"]] = (
    deals_df.loc[
        mask_amount,
        ["Initial Amount Paid (€)", "Offer Total Amount (€)"]
    ].values
)
count = (
    deals_df["Offer Total Amount (€)"]
    < deals_df["Initial Amount Paid (€)"]
).sum()

print(f"Значения в {mask_amount.sum()} строках успешно поменялись местами.")
print(
    f"Количество строк, где Offer Total Amount < Initial Amount Paid: {count}"
)
```

```python
mask_education = (
    (deals_df["Offer Total Amount (€)"] == 11000) &
    (deals_df["Product"] == "UX/UI Design") &
    (
        deals_df["Education Type"].isin(["", "Evening"]) |
        deals_df["Education Type"].isna()
    )
)
deals_df.loc[mask_education, "Education Type"] = "Morning"
```

## 1. Рассчитайте сводную статистику
 (среднее значение, медиана, режим, диапазон) для числовых полей.


```python
def describe_numeric_data(df_dict):
    for df_name, df in df_dict.items():
        numeric_df = df.select_dtypes(include=['number'])

        if numeric_df.empty:
            print(f"⚠️ В таблице {df_name} нет числовых данных.")
            continue

        stats = pd.DataFrame({
            "Среднее": numeric_df.mean(),
            "Медиана": numeric_df.median(),
            "Режим": numeric_df.mode().iloc[0] if not numeric_df.mode().empty else None,
            "Мин": numeric_df.min(),
            "Макс": numeric_df.max(),
            "Диапазон": numeric_df.max() - numeric_df.min()
        }).round(2)

        print(f"📊 Сводная статистика для {df_name}:")
        print(stats.to_string())
        print("\n" + "=" * 80 + "\n")

tables = {
    "processed_deals_df": processed_deals_df,
    "processed_spend_df": processed_spend_df,
    "processed_calls_df": processed_calls_df,
    "processed_contacts_df": processed_contacts_df
}

describe_numeric_data(tables)
```

## 2. Анализируйте категориальные поля,
 такие как качество, стадия, источник и

```python
def analyze_categorical_columns(df, columns):
    for col in columns:
        if col in df.columns:
            print(f"{col} (Уникальных значений: {df[col].nunique()})")
            print("" + "-" * 33 + "\n")
            abs_counts = df[col].value_counts()
            total = abs_counts.sum()
            for category, count in abs_counts.items():
                percentage = (count / total) * 100 if total != 0 else 0.0
                print(f"  - {category}: \033[1;30m{percentage:.2f}% ({count})\033[0m")
        else:
            print(f"\n\033[1;31m{col} - Колонка отсутствует в DataFrame\033[0m")
        print("\n" + "=" * 80 + "\n")

categorical_columns = ['Quality', 'Stage', 'Source', 'Product']
analyze_categorical_columns(processed_deals_df, categorical_columns)
```
